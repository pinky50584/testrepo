## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
!pip install bs4
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

Download the webpage at the url


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data = requests.get(url).text
```

Create a soup object


```python
soup = BeautifulSoup(data, "html.parser")
```

Scrape the `Language name` and `annual average salary`.


```python
table = pd.DataFrame(columns={"No.","Language","Created By","Average Annual Salary","Learning Difficulty"})

for row in soup.find_all('tr'):
    body = row.find_all('td')
    lang = body[1].getText ()
    sal = body[3].getText ()
    print(("{}--->{}").format(lang, sal))
    table = table.append({"Language": lang, "Average Annual Salary": sal}, ignore_index=True)
    table.head()
```

Save the scrapped data into a file named *popular-languages.csv*


```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Writing data to CSV file
popular_language = [['Swift', 130801], ['Python', 114383], ['JAVA', 113865], ['C++', 110981], ['JavaScript', 101013], ['GO', 94082], ['R', 92037], ['C#', 88726], ['SQL', 84727], ['PHP', 84793]]
with open('popular-languages.csv', 'w', newline='') as file:
    csvwriter = csv.writer(file)
    csvwriter.writerow(['Language', 'Performance'])
    for row in popular_language:
        csvwriter.writerow(row)

# Reading data from CSV file using pandas
df = pd.read_csv('popular-languages.csv')

# Plotting
plt.figure(figsize=(16, 7))
plt.bar(df['Language'], df['Performance'], align='center', alpha=0.5)
plt.title('Programming Language')
plt.xlabel('Language')
plt.ylabel('Performance')
plt.xticks(rotation=45)
plt.show()
```
